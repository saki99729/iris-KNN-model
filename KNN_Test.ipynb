```python
import pandas as pd
```

```python
#Import dataSet
dataset = pd.read_csv('iris.csv').values
```

```python
print(dataset,dataset.shape)
```

```python
data = dataset[:,0:4]
print(data.shape)
```

```python
target = dataset[:,4]
print(target.shape)
```

```python
#Separate Data set tarining and testing 20% to 80%
from sklearn.model_selection import train_test_split

train_data,test_data,train_target,test_target=train_test_split(data,target,test_size=0.2)
```

```python
print(test_data.shape)
```

```python
#import KNN
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier()
```

```python
model.fit(train_data,train_target) #train data import
```

```python
predicted_target = model.predict(test_data)
print(predicted_target)
```

```python
print(test_target)
```

```python
from sklearn.metrics import accuracy_score
acc = accuracy_score(test_target,predicted_target)
print(acc)
```

```python
predicted_train_target = model.predict(train_data)
print(predicted_train_target)
```

```python
acc = accuracy_score(train_target,predicted_train_target)
print(acc)
```

```python
import joblib
joblib.dump(model,'iris_model.sav')
```
